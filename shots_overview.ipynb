```python
# Process the first directory

def get_shots_ipsip(work_path):
    """
    Returns a dictionary of form:
        {shotnr: diag_name}

    List the diagnostics (time series data) found in this directory
    
    """
    # Find all shot files, xxxxxx.txt, in this directory
    shots = [int(x[:6]) for x in listdir(work_path) if x.endswith("txt")]
    # Construct to predictor name from the directory structure. In this case it's just ipsip.
    diag_name = "ipsip"
    # Create a dictionary with avialable predictors for each shot found.
    shot_dict = {}
    for shotnr in shots:
        shot_dict.update({shotnr: "ipsip"})
    return shot_dict


shots_ipsip = get_shots_ipsip(path_list[0])
print(shots_ipsip)
```
